## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
import re 
import collections
from collections import defaultdict 

class BPETokenizer:
    '''
    Byte-Pair Encoding Tokenizer

    The tokenizer uses BPE algorithm for tokenization 
    based on https://arxiv.org/pdf/1508.07909.pdf
    '''

    def __init__(self, vocab_size: int) -> None:
        '''
        Initialize BPE tokenizer: the vocab size, 
        corpus (used for stats and merge functions) and vocab
        '''
        self.vocab_size = vocab_size
        self.corpus = defaultdict(int) 
        self.vocab = []


    def get_init_values(self, corpus: List[str]):
        '''
        Fill in a dictionary of words with their frequency in the data (corpus),
        fill in the initial vocab 
        '''
        # Fill the corpus for further functions
        for line in corpus: 
            for word in line.split(): 
                self.corpus[' '.join(list(word)) + ' </w>'] += 1
                
        # Fill the vocab with unique tokens (letters)
        for line in self.corpus:
            self.vocab += line.split()
        self.vocab = list(set(self.vocab))


    def get_stats(self, corpus):
        '''
        Return frequency counts of the pairs in vocabulary
        
        Implementation is taken from 
        https://arxiv.org/pdf/1508.07909.pdf
        '''
        pairs = collections.defaultdict(int)
        for word, freq in corpus.items():
            symbols = word.split()
            for i in range(len(symbols) - 1):
                pairs[symbols[i], symbols[i+1]] += freq
        return pairs
        

    def merge_vocab(self, pair, v_in):
        '''
        Return a new vocabulary with the merged pair
        
        Implementation is taken from 
        https://arxiv.org/pdf/1508.07909.pdf
        '''
        v_out = {}
        bigram = re.escape(' '.join(pair))
        p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        for word in v_in:
            w_out = p.sub(''.join(pair), word)
            v_out[w_out] = v_in[word]
        return v_out

        
    def train(self, corpus: List[str]) -> None:
        '''
        Learn merges and build the vocabulary of the specified vocab_size

        Algorithm implementation is taken from 
        https://arxiv.org/pdf/1508.07909.pdf
        '''
        # Create initial corpus
        self.get_init_values(corpus)

        # Fill the vocab of desired size
        while len(self.vocab) < self.vocab_size: 
            pairs = self.get_stats(self.corpus)
            best = max(pairs, key=pairs.get)
            self.vocab.append(best[0] + best[1])
            self.corpus = self.merge_vocab(best, self.corpus)


    def tokenize(self, text: str) -> List[str]:
        '''
        Tokenize the given data using the pretrained vocabulary
        '''
        tokens = []
        for word in text.split():
            word = word + '</w>'
            while word != '':
                for i in range(len(word)):
                    if word[:len(word)-i] in self.vocab:
                        tokens.append(word[:len(word)-i])
                        break
                word = word[len(word)-i:]
        return tokens
        
```

```python
# Read tweets dataset from the files

tweets = []
for i in range(1, 6):
    path = 'data/file' + str(i)
    with open(path, 'r') as f:
        tweets += f.readlines()
```

```python
# Example

tokenizer = BPETokenizer(2000)
tokenizer.train(tweets)

text = "@dialloc congrats on finding your way over. it may be slow going at first. hang in there. it's kinda cool when u get up to speed."
tokenizer.tokenize(text)
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import matplotlib.pyplot as plt

vocab_sizes = [250, 500, 750, 1000]
avg_lengths = []

for vocab_size in vocab_sizes:
    tokenizer = BPETokenizer(vocab_size)
    tokenizer.train(tweets)
    lengths = [len(tokenizer.tokenize(tweet)) for tweet in tweets]
    avg_lengths.append(sum(lengths) / len(lengths))

plt.plot(vocab_sizes, avg_lengths)
plt.title("Dependency of the average length on vocabulary size")
plt.xlabel("Vocabulary size")
plt.ylabel("Avg number of tokens")
plt.grid(True)
plt.show()
```

On average, increasing vocabulary size decreases the number of tokens in the text as the tokenizer recognizes more lexical items. For this task, I would prefer a vocab_size of 1000, as it tokenizes the text most accurately.
